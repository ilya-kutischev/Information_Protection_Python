```python
#7 лаба
def fpow(base, exp,mod):
    """ Fast power calculation using repeated squaring """
    if exp < 0:
        return 1 / power(base, -exp)
    ans = 1
    while exp:
        if exp & 1:
            ans *= base
            ans = ans % mod
        exp >>= 1
        base *= base
        base = base % mod
    return ans
```

```python
#1 задание и 2 #####################################################################

# декоратор для счёта времени действия функции    
from functools import wraps
from time import time
import random

def timing(f):
    @wraps(f)
    def wrap(*args, **kw):
        ts = time()
        result = f(*args, **kw)
        te = time()
        print ('func:%r args:[%r, %r] took: %2.4f sec' % (f.__name__, args, kw, te-ts))
        return result
    return wrap


def R_b(b,n):
    return pow(b,2,n)


#решето эратосфена для подсчёта простых чисел до n
from math import sqrt, gcd,exp, log
import numpy as np

def sieve_eratosthenes(n):
    n = int(n)
    numbers = list(range(2, n + 1))
    for number in numbers:
        if number != 0:
            for candidate in range(2 * number, n+1, number):
                numbers[candidate-2] = 0    
    return(list(filter(lambda x: x != 0, numbers)))


@timing
def dixon_factor(n,c):
  
    # составляем базу простых по с
    if c>=0.5 and c<=1:
        base = []
        L_n = (exp((log(n)*log(log(n)))**0.5))**c
        base = sieve_eratosthenes(L_n)
        
    else:
        raise ValueError
  
    # извлекаем корень
    start = int(sqrt(n))
  
    # переменная для соответствующих квадратов
    pairs = []
  
    # цткл от корня n до n
    for i in range(start, n):
  
        # находим соответствующие квадраты 
        for j in range(len(base)):
            lhs = fpow(i,2, n)
            rhs = fpow(base[j],2, n)
              
            #если подходят добавляем в массив
            if(lhs == rhs):
                pairs.append([i, base[j]])
  
    new = []
  
    # для каждой пары вычисляем НОД 
    for i in range(len(pairs)):
        factor = gcd(pairs[i][0] - pairs[i][1], n)
          
        # если находим делитель не 1 добавляем в конечный список делителей
        if(factor != 1):
            new.append(factor)
  
    x = np.array(new)
    
    
    # тестируем гладкость
    def test_smooth(a,base):
#         print("testing smooth of ",a)
#         print(base)
        for prime in base:
            while(a%prime==0):
                a = a//prime
#                 print("a=",a)
        if a==1:
            return True
        else:
            return False
        
        
    #возвращает список степеней Б-гладких чисел    
    def return_degree(a,base):
        array = [0]*len(base)
        for i in range(len(base)):
            
            while(a%base[i]==0):
                a = a//base[i]
                array[i]+=1
#                 print("a=",a)
        if a==1:
            return array
    #процедура, методом пробных делений вычисляющая случайные числа
#   b1, b2, …, bk, k = m+ 1, для которых R(bi) является B-гладким числом. 


    @timing
    def proc_ret_b_smooth(base):
        k = len(base)+1
        b=[]
#         print("k=",k)
        while len(b)<k:
            num = random.randrange(2,n**0.5//1)
            Rb_i = R_b(num,n)
            if test_smooth(Rb_i,base):
                b.append([num,return_degree(num,base)])
#             print(b)
        return b
        
        
        
    print("array of b_i which RBi is smooth to base = ",proc_ret_b_smooth(base))   
        
        
    # возвращаем без повторов список найденных делителей
    return(np.unique(x))
```
