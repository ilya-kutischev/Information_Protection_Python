```python
import numpy as np
import time
import os
def naivny(a,d,n):
    b = 1
    for i in range(d):
        b = (a*b)%n
    return b

start = time.time()
print(naivny(5,10**7,100))
end = time.time()
print(end - start)

# 5**(10**7) - 0.5 сек
```

```python
def fast(a,d,n):
    d_bin = int(bin(d)[2:])
    
    b = a
    d_temp = d_bin
    
    k = len(str(d_bin))
    
    arr = []
    
    while d_temp > 0:
        arr.append((d_temp % 10))
        d_temp = d_temp // 10
    #arr = arr[::-1]
    
    
    for i in range(k-1):
        
        b=((b**2)*(a**(arr[i])))%n
    return b

start = time.time()
print(fast(5,10**12000,100))
end = time.time()
print(end - start)

# 5**(10**12000) - 0.5 sec - разница очевидна.
```

```python
def fast_pow(x, y, n): # тот же алгоритм но через рекурсию уже работает плохо
    if y == 0:
        return 1
    if y == -1:
        return 1. / x
    p = fast_pow(x, y // 2, n)
    p *= p
    if y % 2:
        p = (p*x) 
    p = p % n
    return p


start = time.time()
print(fast_pow(5, 10**7, 100))
end = time.time()
print(end - start)
```

```python
# ВЫчисление обратных элементов
def adv_alg_euclid(a,b):
    M = np.array([[1,0],[0,1]])
    while True:
        r = a % b
        q = a // b
        if r == 0:
            gcd = M[0][1]*a+M[1][1]*b
            if gcd == 1:
                return ((M[0][1]%b+b)%b)
            else:
                return -1
        else:
            a = b
            b = r
            M = np.dot(M , np.array([[0,1],[1,-q]]))

start = time.time()             
print(adv_alg_euclid(876233513322246749070,876233513322246749069))
end = time.time()
print(end - start)
```
